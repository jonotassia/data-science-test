```python
# 2nd method
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x, y); # add some data
```

## Matplotlib example workflow

```python
# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
        xlabel="x-axis",
        ylabel="y-axis"
)

# 5. Save and show figure
fig.savefig("./plot.png")
```

## Making Figures with NumPy Arrays

Common examples:
* Line plot
* Scatter plot

```python
# Create some data
x = np.linspace(0, 10, 100)
y = x**2
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(xlabel="x-axis", ylabel="y-axis");
```

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, y);
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from a dictionary
nut_butter_prices = {
    "Almond Butter": 10,
    "Peanut Butter": 8,
    "Cashew Butter": 12
}

fig, ax = plt.subplots()
ax.bar(x=list(nut_butter_prices.keys()), height=list(nut_butter_prices.values()))
ax.set(xlabel="Nut Butters", ylabel="Units Sold", title="Nut Sales Details");
```

```python
# Horizontal bar grapg
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make a histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## Which is the right way to do this (pyplot vs matplotlib OO method)?
* When plotting something quickly, use pyplot method
* When plotting something complex, use matplotlib OO method


```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Hybrid method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax)
ax.set(xlabel="Age", ylabel="Cholesterol", title="Implications of Age on Cholesterol")
ax.set_xlim([45, 100])
```

```python
# OO Method
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);
ax.set(xlabel="Age", ylabel="Cholesterol", title="Implications of Age on Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target");
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
# Subplots thalach, chol, and age - OO Method
fig, (ax0, ax1) = plt.subplots(nrows=2, figsize=(10,10))

# add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"]);

# customize ax0
ax0.set(xlabel="Age", ylabel="Max Heart Rate", title="Implications of Age on Max Heart Rate");
ax0.legend(*scatter.legend_elements(), title="Target");
ax0.axhline(over_50["thalach"].mean(), linestyle='--');

# add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

# add data to ax1
ax1.set(xlabel="Age", ylabel="Cholesterol", title="Implications of Age on Cholesterol");
ax1.legend(*scatter.legend_elements(), title="Target");
ax1.axhline(over_50["chol"].mean(), linestyle='--');

# add title to full plot
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
plt.style.use('seaborn-white')

# Subplots thalach, chol, and age - OO Method
fig, (ax0, ax1) = plt.subplots(nrows=2, sharex=True, figsize=(10,10))

# add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter");

# customize ax0
ax0.set(ylabel="Max Heart Rate", title="Implications of Age on Max Heart Rate");
ax0.legend(*scatter.legend_elements(), title="Target");
ax0.axhline(over_50["thalach"].mean(), linestyle='--', c="red");

# add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter");

# add data to ax1
ax1.set(xlabel="Age", ylabel="Cholesterol", title="Implications of Age on Cholesterol");
ax1.legend(*scatter.legend_elements(), title="Target");
ax1.axhline(over_50["chol"].mean(), linestyle='--', c="red");

# customizing y and x limitations
ax0.set_ylim([60, 200])
ax1.set_ylim([140, 400])
ax1.set_xlim([50, 80])  # this is inherited because they are sharing an axis


# add title to full plot
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
