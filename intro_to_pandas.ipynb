## Manipulating Data

```python
car_sales["Make"] = car_sales["Make"].str.lower()
car_sales
```

```python
car_missing = pd.read_csv("https://cdn.fs.teachablecdn.com/l1WAsummSbaPBpmq4nlR")
car_missing.head()
```

```python
car_missing.fillna("0", inplace=True)
```

```python
car_missing
```

```python
# Column from series
seats_column = pd.Series([5,5,5,5,5])

# New column called seates
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales = car_sales.fillna("5")
car_sales
```

```python
# Column from python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 5, 6, 7, 8, 9]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
# Create a column from other data
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
car_sales
```

```python
# Create a column from a single value
car_sales["Number of Wheels"] = 4
car_sales
```

```python
car_sales = car_sales.drop("Unnamed: 5", axis=1)
car_sales
```

```python
# Shuffling a data frame
car_sales.sample(frac=0.5)
```

```python
# Applying a function
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x/1.6)
car_sales
```

```python
car_sales = car_sales.rename(columns={"Odometer (KM)": "Odometer (Mi)"})
car_sales
```
